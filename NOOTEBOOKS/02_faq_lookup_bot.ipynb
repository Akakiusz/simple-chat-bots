### Dictionary lookup

Where the reflex agent (01) searched for keywords *inside* the input,
this bot matches the **whole input** against a dictionary of known questions.

The trade-off is deliberate:
- Keyword matching is forgiving but imprecise — "track" fires on "I lost my tracking number"
- Exact matching is precise but brittle — "what time do you open?" fails on the question mark

Neither understands language. They fail in opposite directions.



## User Guidance

The chatbot is dedicated to answering frequently asked questions about Bean There Coffee shop.
Use clear and complete phrases for the best results.

#### Supported Queries
The chatbot can handle the following types of questions:
- Greeting (type "hello" or "hi")
- Opening hours (type "what time do you open")
- Closing hours (type "what time do you close")
- Location (type "where are you located")
- WiFi (type "do you have wifi")
- Menu (type "what coffee do you serve")
- Vegan options (type "do you have vegan options")
- Pricing (type "how much is a coffee")
- Payment (type "do you take card")
- To **finish chat** — type "bye"

#### Usage Guidelines
- Use clear and complete phrases
- Avoid abbreviations or slang
- Match the phrasing as closely as possible
- To finish the chat — please use the phrase "bye"

#### Limitations
The chatbot **cannot**:
- Understand partial or rephrased questions
- Provide real-time or personalised information
- Remember previous messages (no memory)
- Learn from interactions

```python
# Importing the random library

import random
```

```python
# Creating a dictionary of customer questions and responses

customer_questions = {
    "what time do you open": "We open at 8am Monday to Friday, and 9am on weekends.",
    "what time do you close": "We close at 6pm every day.",
    "where are you located": "We are located at 12 Main Street, Dublin 1.",
    "do you have wifi": "Yes, free wifi is available for all customers.",
    "what coffee do you serve": "We serve espresso, cappuccino, latte, and americano.",
    "do you have vegan options": "Yes, we have oat milk and a selection of vegan pastries.",
    "how much is a coffee": "A standard coffee starts from 3 euro.",
    "do you take card": "Yes, we accept all major credit and debit cards.",
    "hello": "Hello! Welcome to Bean There Coffee. How can I help you today?",
    "hi": "Hi there! What can I help you with?",        
}   
```

```python
# Creating a list of out of scope responses

out_of_scope_responses = [
    "I'm sorry, I don't understand that question.",
    "Can you please rephrase that?",
    "I'm not sure how to answer that.",
    "That's an interesting question, but I'm not sure how to respond.",
    "I'm sorry, I don't have an answer for that."
]   
```

```python

# Starting the chatbot interaction
print("Bean There Coffee: Hello! Welcome to Bean There Coffee. How can I help you today?")

# Starting a loop to continuously interact with the user until they say goodbye
while True:
    user_input = input("You: ").lower().strip()

    if user_input in ("bye", "goodbye"):
        print("Bean There Coffee: Thanks for visiting! Have a great day!")
        break
# Otherwise, check if the user input is in the customer questions dictionary and respond accordingly
    if user_input in customer_questions:
        print(f"Bean There Coffee: {customer_questions[user_input]}")
    else:
        print(f"Bean There Coffee: {random.choice(out_of_scope_responses)}")
```
