### Model Based Reflex Agent

A simple reflex agent maps the current percept straight to an action.
A model-based agent keeps an internal picture of the world and acts on that.

The difference only matters when a single reading lies.

**The problem:** a cloud passes over the sensor. Light drops for one reading,
then recovers. A reflex agent switches the light on, then off again a second
later. A model-based agent looks at the last few readings and sees that the
world isn't actually dark — one reading was.

Memory here isn't decoration. It's what separates "it's dark" from
"it looked dark for a moment".

```python

# Class definition for a simple reflex agent that turns a light on or off based on the current light level.
class SimpleReflexAgent:
    """Maps the current percept straight to an action."""

    def __init__(self, darkness_threshold=400):
        self.darkness_threshold = darkness_threshold
        self.light_on = False

    def decide(self, light_level):
        return light_level < self.darkness_threshold

    def act(self, light_level):
        should_be_on = self.decide(light_level)
        switched = should_be_on != self.light_on
        self.light_on = should_be_on
        return switched # Some action was taken if the light state changed
```

```python

# Class definition for a model-based reflex agent that acts on the last N readings of light levels, not just the current one.
class ModelBasedReflexAgent:
    """Acts on the last N readings, not just the current one."""

    def __init__(self, darkness_threshold=400, memory_length=5):
        self.darkness_threshold = darkness_threshold
        self.memory_length = memory_length
        self.memory = []
        self.light_on = False

    def perceive(self, light_level):
        self.memory.append(light_level)
        if len(self.memory) > self.memory_length:
            self.memory.pop(0)

# Define the decide method to determine whether to turn the light on or off based on the average of the last N readings.
    def decide(self):  
        if len(self.memory) < self.memory_length:
            return self.light_on
# Return True if all readings are below the darkness threshold, False if all readings are above or equal to the threshold, otherwise return the current state of the light.
        if all(reading < self.darkness_threshold for reading in self.memory):
            return True
        if all(reading >= self.darkness_threshold for reading in self.memory):
            return False
        return self.light_on

# Define the act method to perceive the light level, decide whether to turn the light on or off, and return whether the light state has changed.  
    def act(self, light_level):
        self.perceive(light_level)
        should_be_on = self.decide()
        switched = should_be_on != self.light_on
        self.light_on = should_be_on
        return switched
```

```python
# Test the SimpleReflexAgent and ModelBasedReflexAgent classes with a sequence of light levels.

light_levels = [800, 800, 800, 800, 800,   # clear
                50,                        # cloud
                800, 800, 800, 800,        # sun back
                30,                        # cloud again
                800, 800, 800,             # still bright
                300, 300, 300, 300, 300, 300]  # dusk

def run(agent, readings, label):
    switches = 0
    for level in readings:
        if agent.act(level):
            switches += 1
    print(f"{label:22} {switches} switches")
    return switches

run(SimpleReflexAgent(), light_levels, "Simple reflex agent:")
run(ModelBasedReflexAgent(), light_levels, "Model-based agent:")  
```

### Memory doesn't help?

The cloud scenario is the case memory was built for. But a fair test also asks
where the extra machinery buys nothing.

```python
# Test the SimpleReflexAgent and ModelBasedReflexAgent classes with a different sequence of light levels.

light_levels = [500, 450, 300, 200, 100, 50, 0, 100, 200, 300, 400, 500]

run(SimpleReflexAgent(), light_levels, "Simple reflex agent:")
run(ModelBasedReflexAgent(), light_levels, "Model-based agent:")
```
